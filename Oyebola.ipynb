```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
volt = np.arange(-157.95,0,0.05)
```

```python
volt
```

```python
freq = [((2*math.pi)**-1)*(5*10**-9*((1.3637*10**-3)*(1-v/0.6)**-0.5))**-0.5 for v in volt]
```

```python
fig, ax = plt.subplots()
ax.plot(volt, freq)

ax.set(xlabel='voltage (V)', ylabel='Frequency (Hz)')

ax.grid()
plt.show()
```
